# Conditional Probability — Intuition

Conditional probability answers one basic question:

> If we already know that one event happened, how does that change the probability of another event?

It is the foundation for Bayes' rule, independence, probability trees, medical tests, recommendation systems, and many ML classification problems.


## 1. Basic Idea

Suppose:

- `C` = Applicant has a CS degree
- `F` = Resume is flagged

The probability:

$$P(C)$$

means:

> Probability that a randomly selected applicant has a CS degree.

But:

$$P(C \mid F)$$

means:

> Probability that an applicant has a CS degree, given that the resume was flagged.

The phrase **given that** is the key.


## 2. Formula

Conditional probability is defined as:

$$P(A \mid B) = \frac{P(A \cap B)}{P(B)}$$

This means:

> Restrict attention to the world where `B` happened, then ask what fraction of that world also has `A`.


## 3. Frequency Interpretation

Imagine 1,000 applicants.

- 500 resumes are flagged
- 75 flagged resumes have CS degrees

Then:

$$P(\text{CS Degree} \mid \text{Flagged}) = \frac{75}{500} = 0.15$$

So among flagged resumes, 15% have CS degrees.


```python
total_applicants = 1000
flagged = 500
flagged_and_cs = 75

p_cs_given_flagged = flagged_and_cs / flagged
p_cs_given_flagged
```

## 4. Conditional Probability Is About Changing the Denominator

For ordinary probability:

$$P(A) = \frac{\text{count of A}}{\text{total population}}$$

For conditional probability:

$$P(A \mid B) = \frac{\text{count of A and B}}{\text{count of B}}$$

The denominator changes from the full population to only the group where `B` happened.


```python
total_applicants = 1000
cs_degree = 200
flagged = 500
flagged_and_cs = 75

p_cs = cs_degree / total_applicants
p_cs_given_flagged = flagged_and_cs / flagged

p_cs, p_cs_given_flagged
```

Here:

- $P(\text{CS Degree}) = 20\%$
- $P(\text{CS Degree} \mid \text{Flagged}) = 15\%$

These are different because the denominator is different.


## 5. Conditional Probability vs Bayes' Rule

Conditional probability asks directly:

$$P(A \mid B)$$

Bayes' rule is used when we need to reverse the condition:

$$P(A \mid B) \rightarrow P(B \mid A)$$

Example:

- Given: among flagged resumes, 15% have CS degrees: $P(C \mid F)$
- Asked: among CS degree holders, what percentage are flagged? $P(F \mid C)$

That reversal is where Bayes' rule becomes useful.


## 6. Independence Test

Two events `A` and `B` are independent if knowing `B` does not change the probability of `A`.

$$P(A \mid B) = P(A)$$

If these are not equal, the events are dependent.


```python
p_cs = 0.20
p_cs_given_flagged = 0.15

is_independent = p_cs == p_cs_given_flagged
is_independent
```

Since:

$$P(\text{CS Degree} \mid \text{Flagged}) = 0.15$$

but:

$$P(\text{CS Degree}) = 0.20$$

the two events are **not independent**.


## 7. Generic Helper Function

A simple function for conditional probability using counts:


```python
def conditional_probability(joint_count: int, given_count: int) -> float:
    if given_count == 0:
        raise ValueError("given_count cannot be zero")
    return joint_count / given_count
```

```python
conditional_probability(
    joint_count=75,
    given_count=500,
)
```

## 8. Using a Contingency Table

Conditional probability problems are often easiest with a table.


```python
import pandas as pd

table = pd.DataFrame(
    {
        "CS Degree": [75, 125, 200],
        "No CS Degree": [425, 375, 800],
        "Total": [500, 500, 1000],
    },
    index=["Flagged", "Not Flagged", "Total"]
)

table
```

From the table:

$$P(\text{CS Degree} \mid \text{Flagged}) = \frac{75}{500}$$

$$P(\text{Flagged} \mid \text{CS Degree}) = \frac{75}{200}$$

Notice that these are not the same. The numerator is the same, but the denominator changes.


```python
p_cs_given_flagged = 75 / 500
p_flagged_given_cs = 75 / 200

p_cs_given_flagged, p_flagged_given_cs
```

## 9. Key Takeaways

1. Conditional probability means probability after restricting the sample space.
2. The denominator changes to the event after the `given` symbol.
3. $P(A \mid B)$ and $P(B \mid A)$ are usually different.
4. Independence means $P(A \mid B) = P(A)$.
5. Bayes' rule is used when we need to reverse the conditioning direction.
