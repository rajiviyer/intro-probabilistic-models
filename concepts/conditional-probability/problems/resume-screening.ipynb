# Resume Screening — Conditional Probability and Independence

An AI model flags **50%** of resumes.

**20%** of applicants have a CS degree.

Among flagged resumes, **15%** have a CS degree.

**Question:** Are the events `"has CS degree"` and `"flagged"` independent?


## Step 1: Define the Events

Let:

- `C` = applicant has a CS degree
- `F` = resume is flagged

We are given:

$$P(F) = 0.50$$

$$P(C) = 0.20$$

$$P(C \mid F) = 0.15$$


```python
p_flagged = 0.50
p_cs = 0.20
p_cs_given_flagged = 0.15
```

## Step 2: Independence Rule

Two events `C` and `F` are independent if knowing one event does not change the probability of the other.

So `C` and `F` are independent if:

$$P(C \mid F) = P(C)$$

Here:

$$P(C \mid F) = 0.15$$

but:

$$P(C) = 0.20$$

Since they are not equal, the events are **not independent**.


```python
from probmodels import are_independent

are_independent(
    probability_a=p_cs,
    probability_a_given_b=p_cs_given_flagged,
)
```

## Step 3: Interpret the Result

Among all applicants:

$$P(C) = 20\%$$

Among flagged resumes:

$$P(C \mid F) = 15\%$$

The probability of having a CS degree changes after we know the resume was flagged.

Therefore, the events are **dependent**.


## Step 4: Build a Frequency Table

To make the result more intuitive, imagine **1,000 applicants**.


```python
total_applicants = 1000

flagged_count = int(total_applicants * p_flagged)
cs_count = int(total_applicants * p_cs)

flagged_and_cs_count = int(flagged_count * p_cs_given_flagged)

flagged_and_no_cs_count = flagged_count - flagged_and_cs_count
not_flagged_count = total_applicants - flagged_count
not_flagged_and_cs_count = cs_count - flagged_and_cs_count
not_flagged_and_no_cs_count = not_flagged_count - not_flagged_and_cs_count

flagged_count, cs_count, flagged_and_cs_count
```

```python
import pandas as pd

table = pd.DataFrame(
    {
        "CS Degree": [
            flagged_and_cs_count,
            not_flagged_and_cs_count,
            cs_count,
        ],
        "No CS Degree": [
            flagged_and_no_cs_count,
            not_flagged_and_no_cs_count,
            total_applicants - cs_count,
        ],
        "Total": [
            flagged_count,
            not_flagged_count,
            total_applicants,
        ],
    },
    index=[
        "Flagged",
        "Not Flagged",
        "Total",
    ],
)

table
```

From the table:

$$P(C \mid F) = \frac{75}{500} = 15\%$$

But:

$$P(C) = \frac{200}{1000} = 20\%$$

So the events are not independent.


## Step 5: Optional — Compute the Reverse Conditional Probability

This is where Bayes' rule can enter.

The original question does **not** require Bayes' rule.

But if we ask:

> What fraction of CS degree holders are flagged?

Then we need:

$$P(F \mid C)$$

Using the frequency table:

$$P(F \mid C) = \frac{75}{200} = 37.5\%$$


```python
from probmodels import conditional_probability_from_counts

p_flagged_given_cs = conditional_probability_from_counts(
    joint_count=flagged_and_cs_count,
    given_count=cs_count,
)

p_flagged_given_cs
```

## Final Answer

`Has CS degree` and `flagged` are **not independent**.

Reason:

$$P(C \mid F) \ne P(C)$$

Specifically:

$$0.15 \ne 0.20$$

Knowing that a resume was flagged changes the probability that the applicant has a CS degree.


## Key Lesson

This is mainly a **conditional probability and independence** problem.

It becomes a **Bayes' rule** problem only if we need to reverse the condition, such as going from:

$$P(C \mid F)$$

to:

$$P(F \mid C)$$
