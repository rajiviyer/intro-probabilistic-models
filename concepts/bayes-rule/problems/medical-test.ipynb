# Medical Test - Bayes' Rule

A rare disease affects 1% of people.

A test:

- Detects the disease 95% of the time
- Gives false positives 10% of the time

You test positive.

Question

What is the probability that you actually have the disease?

## Step 1: Define Hypothesis (priors)

```python
priors = {
    "Disease": 0.01,
    "No Disease": 0.99
}
```

## Step 2: Define Likelihoods

```python
likelihoods = {
    "Disease": 0.95,
    "No Disease": 0.10
}
```

## Step 3: Run Bayes Rule

```python
from probmodels import explain_update, simulate_bayes
```

```python
results = explain_update(
    priors, 
    likelihoods,
    population_size=10_000,
    evidence_name="Positive",
    no_evidence_name="Negative",
    draw_tree=True,
    )

results["tree"]
```

## Step 4: Simulate in Python

```python
result = simulate_bayes(
    n=10_000,
    priors=priors,
    likelihoods=likelihoods,
    evidence_name="Positive"
)

result["table"]
```

```python
result["posterior_explanations"]["Disease"]
```

# Observation

#### Strong evidence does not necessarily imply a high posterior probability when the prior probability is very small
