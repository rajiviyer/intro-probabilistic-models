### Problem Statement

In a large tech-savvy population, people either are coders or not. Roughly 30% of them are
coders. Among coders, 80% have a GitHub account. Among non-coders, 20% have a GitHub account
— maybe they created one during college, for a course, or just out of curiosity.
Now suppose you meet a person at random, and you learn that they have a GitHub account.
Based on this information, what is the probability that they are actually a coder?

#### Step 1. Define Hypothesis

```python
priors = {
    "Coders": 0.30,
    "Non-Coders": 0.70
}
```

#### Step 2. Define Likelihoods

```python
likelihoods = {
    "Coders": 0.80,
    "Non-Coders": 0.20 
}
```

```python
from probmodels import bayes_update, explain_update, normalize
```

```python
result = explain_update(
    priors,
    likelihoods,
    population_size=10_000,
    evidence_name="Github",
    no_evidence_name="Non Github",
    draw_tree=True,
)

result["tree"]
```
